```python
#Importing libraries
import numpy as npy
import scipy as spy
import pandas as pd
import sklearn as sk
import matplotlib as plt
```

```python
#1. Read the file and display columns
cruise = pd.read_csv("cruise_ship_info.csv")
cruise
```

```python
#2. Calculate basic statistics of the data (mean, sum, max, min, count, median, std, var) and
#examine data and state your observations.


#Ship_name --> only count can be found 

count_Ship_name = cruise['Ship_name'].count() 
stats_Ship_name = npy.array(['NA','NA','NA','NA',count_Ship_name,'NA','NA','NA'])

#Cruise_line --> only count can be found 

count_Cruise_line = cruise['Cruise_line'].count()
stats_Cruise_line = npy.array(['NA','NA','NA','NA',count_Cruise_line,'NA','NA','NA'])

#Age

mean_Age = cruise['Age'].mean()
sum_Age = cruise['Age'].sum()
max_Age = cruise['Age'].max()
min_Age = cruise['Age'].min()
count_Age = cruise['Age'].count()
median_Age = cruise['Age'].median() 
std_Age = cruise['Age'].std() 
var_Age = cruise['Age'].var()  

stats_Age = npy.array([mean_Age,sum_Age,max_Age,min_Age,count_Age,median_Age,std_Age,var_Age])



#Tonnage

mean_Tonnage = cruise['Tonnage'].mean()
sum_Tonnage = cruise['Tonnage'].sum()
max_Tonnage = cruise['Tonnage'].max()
min_Tonnage = cruise['Tonnage'].min()
count_Tonnage = cruise['Tonnage'].count()
median_Tonnage = cruise['Tonnage'].median() 
std_Tonnage = cruise['Tonnage'].std() 
var_Tonnage = cruise['Tonnage'].var()  

stats_Tonnage = npy.array([mean_Tonnage,sum_Tonnage,max_Tonnage,min_Tonnage,count_Tonnage,median_Tonnage,
                           std_Tonnage,var_Tonnage])


#passengers

mean_passengers = cruise['passengers'].mean()
sum_passengers = cruise['passengers'].sum()
max_passengers = cruise['passengers'].max()
min_passengers = cruise['passengers'].min()
count_passengers = cruise['passengers'].count()
median_passengers = cruise['passengers'].median() 
std_passengers = cruise['passengers'].std() 
var_passengers = cruise['passengers'].var()  

stats_passengers = npy.array([mean_passengers,sum_passengers,max_passengers,min_passengers,count_passengers,
                              median_passengers,std_passengers,var_passengers])

#length

mean_length = cruise['length'].mean()
sum_length = cruise['length'].sum()
max_length = cruise['length'].max()
min_length = cruise['length'].min()
count_length = cruise['length'].count()
median_length = cruise['length'].median() 
std_length = cruise['length'].std() 
var_length = cruise['length'].var()  

stats_length = npy.array([mean_length,sum_length,max_length,
                          min_length,count_length,median_length,std_length,var_length])


#cabins

mean_cabins = cruise['cabins'].mean()
sum_cabins = cruise['cabins'].sum()
max_cabins = cruise['cabins'].max()
min_cabins = cruise['cabins'].min()
count_cabins = cruise['cabins'].count()
median_cabins = cruise['cabins'].median() 
std_cabins = cruise['cabins'].std() 
var_cabins = cruise['cabins'].var()  

stats_cabins = npy.array([mean_cabins,sum_cabins,max_cabins,
                          min_cabins,count_cabins,median_cabins,std_cabins,var_cabins])


#passenger_density

mean_passenger_density = cruise['passenger_density'].mean()
sum_passenger_density = cruise['passenger_density'].sum()
max_passenger_density = cruise['passenger_density'].max()
min_passenger_density = cruise['passenger_density'].min()
count_passenger_density = cruise['passenger_density'].count()
median_passenger_density = cruise['passenger_density'].median() 
std_passenger_density = cruise['passenger_density'].std() 
var_passenger_density = cruise['passenger_density'].var()  

stats_passenger_density = npy.array([mean_passenger_density,sum_passenger_density,
                                     max_passenger_density,min_passenger_density,count_passenger_density,
                                     median_passenger_density,std_passenger_density,var_passenger_density])

#crew

mean_crew = cruise['crew'].mean()
sum_crew = cruise['crew'].sum()
max_crew = cruise['crew'].max()
min_crew = cruise['crew'].min()
count_crew = cruise['crew'].count()
median_crew = cruise['crew'].median() 
std_crew = cruise['crew'].std() 
var_crew = cruise['crew'].var()  

stats_crew = npy.array([mean_crew,sum_crew,max_crew,min_crew,count_crew,median_crew,std_crew,var_crew])



```

```python
"""
6. Create training and testing sets (use 60% of the data for the training and reminder for testing).
"""
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
#specify the target variable (dependent variable = crew) and the independent variables: tonnage,passengers,length,cabins
#crew
x = df_pred_crew.drop(['crew'],axis = 1)
y = df_pred_crew.crew
x
y
```

```python
#split the data set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
#7. Build a machine learning model to predict the ‘crew’ size.
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
## The line / model
plt.pyplot.scatter(y_test, predictions)
plt.pyplot.xlabel('True Values')
plt.pyplot.ylabel('Predictions')

print ('Score: ', model.score(X_test, y_test))
```
